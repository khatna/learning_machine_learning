# Feature Expansion: conference activity in a building

7 is out flow, 9 is in flow

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('CalIt2.csv', header = None)
df.columns = ['flow', 'day', 'time', 'count']
df.head()
```

Let's calculate the number of people in the building at every half-hour by computing the net flow, and taking the cumulative sum

```python
counts = df['count'].to_numpy()
a = np.empty(len(counts))

# Calculating net flow at each half-hour
counts = counts.reshape(counts.shape[0]//2, 2)
counts = counts[:,1] - counts[:,0]
plt.figure(figsize=(10,8))
plt.plot(counts[:1000])

# Cumulative sum
counts = np.cumsum(counts)

plt.plot(counts[:1000])
plt.show()
```
